```python
from __future__ import print_function,division,absolute_import
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#numpy and scipy import
import numpy as np
np.random.seed(1337)
```

```python
from scipy.fftpack import fft,ifft
from sklearn.preprocessing import normalize
from scipy.io import wavfile # get the api
```

```python
filename = 'B_eng_f1.wav'
fs, data = wavfile.read(filename) # load the data

print(data.shape,fs)

plt.plot(data,'c')
plt.title('Data over time (RAW)')
plt.xlabel('Time (in micros)')
plt.ylabel('Amplitude')
plt.show()
```

# Normalizing Waves

```python
data_norm = normalize(data.reshape(1,-1).astype('float32')).ravel()
data_norm.shape

plt.plot(data_norm,'g')
plt.title('Data over time (Normalized)')
plt.xlabel('Time (in micros)')
plt.ylabel('Amplitude')
plt.show()
```

# Let's do some fourier transformation now

```python
data_fft = fft(data)
data_ifft = ifft(data_fft)
data_new = np.absolute(data_fft)

data_new = normalize(data_new.reshape(1,-1).astype('float32')).ravel()

plt.plot(data_new,'r')
plt.title('Data over freq (fft)')
plt.xlabel('Time (in micros)')
plt.ylabel('Amplitude')
plt.show()
```
